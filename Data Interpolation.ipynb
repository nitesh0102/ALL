## Data Interpolation 

```python
Data interpolation is the process of estimating unknown variables based on the known variables within the dataset. 
```

```python
Data interpolation using numpy 
```

```python
1. Linear interpolation 
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,2,3,4,4])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.interp(x_new,x,y)
```

```python
y_interp
```

```python
2. cubic interpolation
```

```python
x = np.array([1,2,3,4,5])
y= np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x,y,kind = 'cubic') 
```

```python
x_new = np.linspace(1,5,10)
y_interp = f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

```python
3. polynomial 
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
f1 = np.polyfit(x,y,2)
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.polyval(f1,x_new)
```

```python
plt.scatter(x_new,y_interp)
```
